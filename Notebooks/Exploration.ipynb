## Importation  des  donnees

### existant

```python
import pandas  as  pd

dpe_existant =  pd.read_csv('../Data/existant_69.csv')
dpe_existant.head()
```

###  neufs

```python
# Lire le fichier CSV situé à '../Data/neufs_69.csv' dans un DataFrame nommé 'dpe_neuf'
dpe_neuf = pd.read_csv('../Data/neufs_69.csv')

# Afficher les premières lignes du DataFrame 'dpe_neuf'
dpe_neuf.head()
```

```python
# Ajouter une colonne 'Logement' avec la valeur 'ancien' au DataFrame 'dpe_existant'
dpe_existant["Logement"] = "ancien"

# Ajouter une colonne 'Logement' avec la valeur 'neuf' au DataFrame 'dpe_neuf'
dpe_neuf["Logement"] = "neuf"
```

```python
# Ajouter une colonne 'Année_construction' avec la valeur 2024 au DataFrame 'dpe_neuf'
dpe_neuf["Année_construction"] = 2024
```

```python
# Trouver les colonnes communes entre les DataFrames 'dpe_existant' et 'dpe_neuf'
common_columns = dpe_existant.columns.intersection(dpe_neuf.columns)

# Convertir l'Index des colonnes communes en liste
common_columns = common_columns.to_list()
```

```python
# Concaténer les DataFrames 'dpe_existant' et 'dpe_neuf' en utilisant uniquement les colonnes communes
# 'join' est défini sur 'inner' pour ne conserver que les colonnes présentes dans les deux DataFrames
# 'ignore_index' est défini sur True pour réinitialiser les index dans le DataFrame concaténé
df_concat = pd.concat([dpe_existant, dpe_neuf], join='inner', ignore_index=True)

# Afficher le DataFrame concaténé
df_concat
```

```python
df_concat['Etiquette_DPE'].value_counts(normalize=True)
```

```python
df_concat["Annee_reception_DPE"] = pd.to_datetime(df_concat['Date_réception_DPE']).dt.year
df_concat["Annee_reception_DPE"].head(10)
```

```python
df_concat['Somme_coûts'] = df_concat['Coût_chauffage'] + df_concat['Coût_éclairage'] + df_concat['Coût_ECS'] + df_concat['Coût_refroidissement'] + df_concat['Coût_auxiliaires']

# Créer une colonne pour vérifier si le coût total correspond à la somme
df_concat['Vérification_coût_total'] = df_concat['Coût_total_5_usages'] == df_concat['Somme_coûts']
```

```python
df_concat["Coût chauffage en %"] = (df_concat['Coût_chauffage']/df_concat['Coût_total_5_usages'])*100
df_concat["Coût chauffage en %"]
```

```python
df_concat['passoire_energetique'] = df_concat['Etiquette_DPE'].isin(['F', 'G'])
df_concat['passoire_energetique'].value_counts(normalize=True)
```

```python
def periode_contruct(annee):
    if annee <= 1960:
        return "Avant 1960"
    if annee >= 1961 and annee <= 1970:
        return "1961 - 1970"
    if annee >= 1971 and annee <= 1970:
        return "1971 - 1980"
    if annee >= 1981 and annee <= 1990:
        return "1981 - 1990"
    if annee >= 1991 and annee <= 2000:
        return "1991 - 2000"
    if annee >= 2000 and annee <= 2010:
        return "2001 - 2010"
    if annee > 2010:
        return "Après 2010"
```

```python
df_concat["Periode_construction"] = df_concat["Année_construction"].apply(periode_contruct)
```

```python
df_concat[["Année_construction","Periode_construction"]].value_counts()
```

```python
df_concat.isnull().mean().sort_values(ascending=True)
```

```python
# Calculer le pourcentage de valeurs manquantes par colonne
missing_percentages = df_concat.isnull().mean().sort_values(ascending=True)

# Afficher le pourcentage de valeurs manquantes pour chaque colonne
for column, percentage in missing_percentages.items():
    print(f"{column}:\t\t\t\t {percentage:.2%}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Exemple : Calculer la corrélation entre deux variables catégorielles
tableau_croise = pd.crosstab(df_concat['N°_DPE_remplacé'], df_concat['Etiquette_DPE'])  # Crée une table de contingence
chi2, p, dof, expected = chi2_contingency(tableau_croise)  # Effectue le test du Chi-Carré

print(f"Valeur de Chi-2 : {chi2}")
print(f"Valeur p : {p}")
```
